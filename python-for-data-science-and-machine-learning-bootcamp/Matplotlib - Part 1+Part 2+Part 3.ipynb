```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# functional method
plt.plot(x,y,'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Larger plot')
axes2.set_title('smaller plot')
```

```python
fig,axes = plt.subplots(1,2)
for cur_ax in axes:
    cur_ax.plot(x,y)

#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
#axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2),dpi=100,facecolor='r',edgecolor='b')
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend(loc=(0.1,0.1))#tuple specifying the location 
```
